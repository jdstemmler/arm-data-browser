```python
import pandas as pd
import datetime
from collections import OrderedDict
```

```python
table = pd.read_csv('../interesting_cases.csv',
                    names=['tstamp', 'case_date', 'description', 'category', 'name', 'email'],
                    parse_dates=[0,],
                    header=0)

table = table[~table['description'].isnull()]
```

```python
table['case_dt'] = table['case_date'].apply(lambda x: datetime.datetime.strptime(x, '%m/%d/%Y'))

table['category'].fillna('Other', inplace=True)
table['category_list'] = table['category'].apply(lambda x: x.split(', '))
categories = set(c.lower() for row in table['category_list'] for c in row)
categories
```

```python
table[['case_dt', 'category_list', 'description']].set_index('category_list')
```

```python
new_table = table.set_index('case_dt')[['description', 'category_list']].sort_index()
new_table
```

```python
grouped = new_table.groupby('category')
sorted_categories = grouped.count().sort_values('description', ascending=False)

d = OrderedDict()
for i, j in sorted_categories.iterrows():
    d[i] = grouped.get_group(i).to_dict()
```
